```python
import seaborn as sns
mytitanic = sns.load_dataset('titanic')
#Loading pre defined data set called titanic
```

```python
#To see all the coln headings in the data set
mytitanic.head()
```

```python
#Distribution plot of fare
sns.distplot(mytitanic['fare'])
```

```python
sns.distplot(mytitanic['fare'],kde = False)
#To remove the kde (gaussian kernel density estimate)
```

```python
sns.jointplot(x='age',y='fare',data=mytitanic)
#Similar to scatterplot
```

```python
#to get hexagons instead of dots in the above plot
sns.jointplot(x='age',y='fare',data=mytitanic,kind = 'hex')
#If kind=reg we get regression
```

```python
#Loading a new pre definded data set from seaborn
mytips = sns.load_dataset('tips')
mytips.head()
```

```python
sns.pairplot(mytips)
```

```python
#Catplot - When we need to plot too many factors
df = sns.load_dataset('exercise')
df.head()
```

```python
plotty = sns.catplot(x = 'time',y  = 'pulse',hue = 'kind',data = df)
```

```python
#Now to get plot on basis of type of diet
plotty2 = sns.catplot(x = 'time',y  = 'pulse',hue = 'kind',col = 'diet',data = df)
```
